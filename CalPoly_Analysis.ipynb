```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import data
csv_16_17 = 'Data/16_17.csv'
csv_17_18 = 'Data/17_18.csv'
csv_18_19 = 'Data/18_19.csv'
csv_19_20 = 'Data/19_20.csv'
csv_20_21 = 'Data/20_21.csv'
```

```python
# Create DataFrames
df_16_17 = pd.read_csv(csv_16_17)
df_17_18 = pd.read_csv(csv_17_18)
df_18_19 = pd.read_csv(csv_18_19)
df_19_20 = pd.read_csv(csv_19_20)
df_20_21 = pd.read_csv(csv_20_21)
```

```python
# Combine the DataFrames
df_all = pd.concat([df_16_17, df_17_18, df_18_19, df_19_20, df_20_21])
```

```python
# Look at the data
df_all.head()
```

```python
# Keep only AggregateLevel = School (gets all school-level data totals with one record per school with school totals)
df_agg = df_all[(df_all['AggregateLevel'] == 'S') & 
                (df_all['CharterSchool'] == 'All') & 
                (df_all['DASS'] == 'All') & (df_all['ReportingCategory'] != 'TA')]
```

```python
# Look at the data
df_agg.columns
```

```python
# Keep only relevant columns
df_cleaned_columns = df_agg[['AcademicYear', 'AggregateLevel', 'CountyCode', 'DistrictCode',
       'SchoolCode', 'CountyName', 'DistrictName', 'SchoolName',
       'ReportingCategory', 'CohortStudents',
       'Regular HS Diploma Graduates (Count)',
       'Regular HS Diploma Graduates (Rate)', "Met UC/CSU Grad Req's (Count)",
       "Met UC/CSU Grad Req's (Rate)", 'Seal of Biliteracy (Count)',
       'Seal of Biliteracy (Rate)']]
```

```python
# Look at the data
with pd.option_context('display.max_rows', 10, 'display.max_columns', None):
    display(df_cleaned_columns)
```

```python
# Keep only rows where 'Met UC/CSU Grad Req's (Count)' is greater than 0
df_met_req = df_cleaned_columns.loc[df_agg["Met UC/CSU Grad Req's (Count)"] > '0']
with pd.option_context('display.max_rows', 10, 'display.max_columns', None):
    display(df_met_req)
```

```python
# Export to CSV
df_met_req.to_csv(r'cleaned_data.csv', index=False)
```
